```python
import cv2
image_path = "/data3/data/18112025/10.143.64.98_8000_35_11_-1/20251118_20251118150459_20251118161903_150501.jpg"   # <-- change this
image = cv2.imread(image_path)

if image is None:
    print("Image not found!")
    exit()

clone = image.copy()

roi = cv2.selectROI("Select Timestamp Region", image, fromCenter=False, showCrosshair=True)

x, y, w, h = roi
print("Selected Coordinates:")
print(f"x_start: {x}")
print(f"y_start: {y}")
print(f"x_end: {x + w}")
print(f"y_end: {y + h}")

cv2.rectangle(clone, (x, y), (x + w, y + h), (0, 0, 255), 2)
cv2.imshow("Selected Region", clone)


masked = image.copy()
masked[y:y+h, x:x+w] = 0

cv2.imshow("Masked Output", masked)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import os
from moviepy import VideoFileClip
from datetime import timedelta

def get_total_video_duration(directory):
    total_seconds = 0

    video_extensions = ('.mp4', '.mkv', '.avi', '.mov', '.flv', '.wmv')

    if not os.path.exists(directory):
        print(f"Directory '{directory}' not found.")
        return 0

    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.lower().endswith(video_extensions):
                file_path = os.path.join(root, file)
                try:
                    with VideoFileClip(file_path) as video:
                        duration = video.duration  
                        print(f"{file}: {timedelta(seconds=int(duration))}")
                except Exception as e:
                    print(f"Error reading {file}: {e}")

    return total_seconds

if __name__ == "__main__":
    folder_path = "chunks" 
    total_duration = get_total_video_duration(folder_path)

    print("\nTotal Duration:")
    print(str(timedelta(seconds=int(total_duration))))
```
